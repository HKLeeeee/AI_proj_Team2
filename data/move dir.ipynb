```python
import os
import shutil
```

```python
mask_dir = './masks'
masks = os.listdir(mask_dir)
masks.remove('.ipynb_checkpoints')

img_dir = './train_images'
img = os.listdir(img_dir)

mask_num = len(masks)
print(mask_num)
mask_file_name = [m[:-4] for m in masks]
masked_img = list(filter(lambda x : x[:-4] in mask_file_name, img))
print(len(masked_img))
```

```python
for idx in range(valid_idx) :
    mask_src = os.path.join(mask_dir, masks[idx])
    mask_dst = os.path.join('./segmentation/trainannot', masks[idx])
    shutil.copyfile(mask_src, mask_dst)
    
    img_src = os.path.join(img_dir, masked_img[idx])
    img_dst = os.path.join('./segmentation/train', masked_img[idx])
    shutil.copyfile(img_src, img_dst)
    
for idx in range(valid_idx, mask_num) :
    src = os.path.join(mask_dir, masks[idx])
    mask_dst = os.path.join('./segmentation/validannot', masks[idx])
    shutil.copyfile(mask_src, mask_dst)
    img_src = os.path.join(img_dir, masked_img[idx])
    img_dst = os.path.join('./segmentation/validation', masked_img[idx])
    shutil.copyfile(img_src, img_dst)
```
