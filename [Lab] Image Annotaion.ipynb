[참고](https://opencv-python.readthedocs.io/en/latest/doc/09.imageThresholding/imageThresholding.html)


```python
test = cv2.imread('./data/train/CASE01/image/CASE01_01.png', 0)
test_c = cv2.imread('./data/train/CASE01/image/CASE01_01.png')
fig = plt.figure()
ax1=fig.add_subplot(1,2,1)
ax1.imshow(test, cmap='gray')
ax2=fig.add_subplot(1,2,2)
ax2.imshow(test_c)
plt.show()
print(test.shape)
print(test_c.shape)
```

```python
ret, thresh1 = cv2.threshold(test,0,255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(test,0,255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(test,0,255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(test,0,255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(test,0,255, cv2.THRESH_TOZERO_INV)
```

```python
fig =  plt.figure(figsize=(10,6))
for i, img in enumerate([thresh1, thresh2, thresh3, thresh4, thresh5]):
    ax = fig.add_subplot(2,3,i+1)
    ax.imshow(img)
    ax.set_title('thresh'+str(i+1))
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
ret, thresh1 = cv2.threshold(test,0,255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(test,0,255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret, thresh3 = cv2.threshold(test,0,255, cv2.THRESH_OTSU)
```

```python
fig =  plt.figure(figsize=(10,6))
for i, img in enumerate([thresh1, thresh2, thresh3]):
    ax = fig.add_subplot(1,3,i+1)
    ax.imshow(img)
    ax.axis('off')
    ax.set_title('thresh'+str(i+1))

plt.tight_layout()
plt.show()
```

```python
# Image File만 따로 모음
import os
target_directory = './data/train_images'
os.mkdir(target_directory)
```

```python
cases = os.listdir('./data/train')
print(len(cases))
```
